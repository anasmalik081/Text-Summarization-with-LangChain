### PromptTemplate

```python
from langchain import PromptTemplate
from langchain.chains import LLMChain
```

```python
template = (
    """
    Write a summary of the following speech:

    Speech: {speech}
    Translate the precise summary to {language}
    """
)
prompt = PromptTemplate(
    input_variables=["speech", "language"],
    template=template
)
```

```python
chain = LLMChain(llm=model, prompt=prompt)
summary = chain.run({"speech": speech, "language": "Hindi"})
```

### Stuff Document Chain Text Summarization

```python
from langchain_community.document_loaders import PyPDFLoader
from langchain.prompts import PromptTemplate
```

```python
loader = PyPDFLoader("apjspeech.pdf")
documents = loader.load_and_split()
documents
```

```python
template = (
    """
    Write a concise and short summary of the following speech.
    
    Speech: {text}
    """
)
prompt = PromptTemplate(
    input_variables=["text"],
    template=template
)
```

```python
from langchain.chains.summarize import load_summarize_chain
```

```python
chain = load_summarize_chain(model, chain_type="stuff", prompt=prompt, verbose=True)
```

```python
output_summary = chain.invoke(documents)
output_summary
```

### Map-Reduce to Summarize Large Documents

```python
from langchain.text_splitter import RecursiveCharacterTextSplitter
```

```python
model = init_chat_model("gemma2-9b-it", model_provider="groq")
```

```python
text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=2000, chunk_overlap=100
)
final_documents = text_splitter.split_documents(documents=documents)
```

```python
final_documents
```

```python
chunk_template = (
    """
    Please summarize the below speech:

    Speech: {text}
    Summary:
    """
)
map_prompt = PromptTemplate(
    input_variables=["text"],
    template=chunk_template
)
```

```python
combine_template = """
Provide the final summary of the entire speech with these important points.
- Add a motivation title
- Start the precise summary with an introduction
- Provide the summary in nmber points for the speech

Summaries: {text}
Final Summary:
"""

combine_prompt = PromptTemplate(
    input_variables=["text"],
    template=combine_template
)
```

```python
summary_chain = load_summarize_chain(
    llm=model,
    chain_type="map_reduce",
    map_prompt=map_prompt,
    combine_prompt=combine_prompt,
    verbose=True
)
```

```python
output_summary = summary_chain.run(final_documents)
```

```python
output_summary
```

### Refine Chain Summarization

```python
chain = load_summarize_chain(
    llm=model,
    chain_type="refine",
    verbose=True
)
output_summary = chain.run(final_documents)
output_summary
```
